```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
totalImages = 0

for i in range(1000):
    
    url = "https://en.wikipedia.org/wiki/Special:Random"
    html = requests.get(url).text
    soup = BeautifulSoup(html, "html.parser")

    for tag in soup.find_all():
        if tag.name == "img":
            totalImages +=1

print(totalImages/1000)
```

```python
url = "https://www.scrapethissite.com/pages/simple/"
html = requests.get(url).text
soup = BeautifulSoup(html, "html.parser")

countries = soup.select(".country-name")
capitals = soup.select(".country-capital")
population = soup.select(".country-population")
area = soup.select(".country-area")
df = pd.DataFrame()

countriesStrs = []
capitalsStrs = []
populationStrs = []
areaStrs = []
for i in countries:
    countriesStrs.append(i.text)

for i in capitals:
    capitalsStrs.append(i.text)

for i in population:
    populationStrs.append(i.text)

for i in area:
    areaStrs.append(i.text)
    
for i in range(len(countries)):
    countriesStrs[i] = countriesStrs[i].strip()
    capitalsStrs[i] = capitalsStrs[i].strip()
    populationStrs[i] = populationStrs[i].strip()
    areaStrs[i] = areaStrs[i].strip()

df["countries"] = countriesStrs
df["capitals"] = capitalsStrs
df["population"] = populationStrs
df["areaStrs"] = areaStrs

df.head()
```
